```python
#look at correlation betwe numeric values 

numeric_df = df.select_dtypes(include=['number'])
numeric_df.corr()

#correlations are really high close to 1 so almost exactly correlated between countries cuz for most countries pop. is increasing steadily w/ each year 

#area corr is low cuz sometimes fee countires with reall high pop and small area or vise versa so no 1:1 corr there 

#The closer r is to zero, the weaker the linear relationship.
#Positive r values indicate a positive correlation, where the values of both variables tend to increase together.
#Negative r values indicate a negative correlation, where the values of one variable tend to increase when the values of the other variable decrease.
```

```python
numeric_df = df.select_dtypes(include=['number'])
sns.heatmap(numeric_df.corr(), annot = True)  #can change colorcoding using cmap='coolwarm' for ex #annot da bi7ot il kalaam gowa il table

plt.rcParams['figure.figsize'] = (20,7)
#or the below
#plt.figure(figsize=(3,10))

plt.show()

#can see strong population icnreasing every year, but area not really corr to pop size
```
